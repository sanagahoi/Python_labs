Given an array A of length N. Find the number of ordered pairs (i, j) which satisfy the given condition

(A[i] - A[j])/(i-j) =1

A[i] -A[j] Should be divisible by i-j and

i != j

````python
def find_pair(arr, N):
    count = 0
    for i in range(N):
        for j in range(i+1, N):
            if arr[i] - arr[j] == i-j:
                count = count + 1
            if arr[j] - arr[i] == j-i:
                count += 1
    return count


A = [1,2,4,3,5,11,34,35,22,32,5]
N  = len(A)
result = find_pair(A, N)
print(result)

def find_ordered_pairs(arr, n):
  count = 0
  i =0
  while i < n:
    j= n-1-i
    if A[i] - A[j] == i-j:
      count = count + 1
    if A[j] - A[i] == j-i:
      count += 1
    i +=1

  return count

N = int(input())
A = list(map(int, input().split()))
result = find_ordered_pairs(A, N)
print(result)
````

````python
# upper to lower case without using inbuilt functions
def to_lower(input_string):
    result = []
    for char in input_string:
        if 'A' <= char <= 'Z':
            result.append(chr(ord(char) + 32))
        else:
            result.append(char)
    return ''.join(result)

text = "Hello World"
print(f"Original: {text}")
print(f"Lowercase: {to_lower(text)}")
````

````python
# 2 sum or 3 sum
# Two Sum problem: Find two numbers in an array that add up to a specific target.
def two_sum(nums, target):
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    return []

nums = [2, 7, 11, 15]
target = 9
result = two_sum(nums, target)
print(f"Numbers: {nums}, Target: {target}, Indices: {result}")

nums = [3, 2, 4]
target = 6
result = two_sum(nums, target)
print(f"Numbers: {nums}, Target: {target}, Indices: {result}")
````

````python
class Hashtable2:
  def __init__(self):
    self.Max = 100
    self.arr = [[] for i in range(self.Max)]

  def get_hash(self, key):
    h=0
    for char in key:
      h+=ord(char)
    return h%self.Max

  def __getitem__(self, index):
    h = self.get_hash(index)
    for element in self.arr[h]:
      if element[0] == index:
        return element[1]

  def __setitem__(self, key, val):
    h = self.get_hash(key)
    found = False
    for idx, element in enumerate(self.arr[h]):
      if len(element)==2 and element[0]==key:
        self.arr[h][idx] = (key, val)
        found = True
        break
    if not found:
      self.arr[h].append((key, val))

  def __delitem__(self, key):
    h = self.get_hash(key)
    for index, element in enumerate(self.arr[h]):
      if element[0] == key:
        del self.arr[h][index]

u = Hashtable2()
u['march 6'] = 120
u['march 6'] = 133
u['march 7'] = 122
u['march 10'] =200
u['march 16'] =300
del u['march 16']
````

Implement doubly linked list.

The only difference with regular linked list is that double linked has prev node reference as well.

That way you can iterate in forward and backward direction. Your node class will look this,



```

class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
```

````python
# doubly linked list

class Node:
  def __init__(self, data=None, next=None, prev=None):
    self.data = data
    self.next = next
    self.prev = prev

class DoublyLinkedList:
  def __init__(self):
    self.head = None

  def print_forward(self):
    # This method prints list in forward direction. Use node.next
    if self.head is None:
      print("list is Empty")
      return

    itr = self.head
    llstr = ''
    while itr:
      llstr += str(itr.data) + '-->'
      itr = itr.next
    print(llstr)


  def print_backward(self):
    # Print linked list in reverse direction. Use node.prev for this.
    if self.head is None:
      print("linked list is Empty")
      return

    itr= self.head
    llstr = ''
    while itr.next:
      itr = itr.next
    while itr:
      llstr += str(itr.data) + '-->'
      itr = itr.prev
    print("Link list in reverse: ", llstr)

  def get_length(self):
    # return length of the linked list in constant time
    if self.head is None:
      return 0

    itr = self.head
    count = 0
    while itr:
      count+=1
      itr = itr.next
    return count

  def insert_at_begining(self, data):
    # Inserting at beginning
    if self.head == None:
      node = Node(data, self.head, None)
      self.head = node
    else:
      node = Node(data, self.head, None)
      self.head.prev = node
      self.head = node

  def insert_at_end(self, data):
    # Inserting at end
    if self.head is None:
      self.head = Node(data, None, None)
      return

    itr = self.head
    while itr.next:
      itr = itr.next
    itr.next = Node(data, None, itr)

  def insert_at(self, index, data):
    # Insert at a given index
    if index<0 or index>self.get_length():
      raise Exception("Invalid Index")

    if index==0:
      self.insert_at_begining(data)
      return

    count = 0
    itr=self.head
    while itr:
      if count == index - 1:
        node = Node(data, itr.next, itr)
        if node.next:
          node.next.prev = node
        itr.next = node
        break
      itr = itr.next
      count += 1

  def remove_at(self, index):
    if index<0 or index>=self.get_length():
      raise Exception("Invalid Index")

    if index==0:
      self.head = self.head.next
      self.head.prev = None
      return

    count = 0
    itr = self.head
    while itr:
      if count == index:
        itr.prev.next = itr.next
        if itr.next:
          itr.next.prev = itr.prev
        break
      itr = itr.next
      count+=1

  def insert_values(self, data_list):
    self.head = None
    for data in data_list:
      self.insert_at_end(data)

if __name__ == '__main__':
  ll = DoublyLinkedList()
  ll.insert_values(["banana","mango","grapes","orange"])
  ll.print_forward()
  ll.print_backward()
  ll.insert_at_end("figs")
  ll.print_forward()
  ll.insert_at(0,"jackfruit")
  ll.print_forward()
  ll.insert_at(6,"dates")
  ll.print_forward()
  ll.insert_at(2,"kiwi")
  ll.print_forward()
````

````python
# binary search tree demonstration

class BinarySearchTreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

  def add_child(self, data):
    if data == self.data:
      return
    if data < self.data:
      # add data in left sub tree
      if self.left:
        self.left.add_child(data)
      else:
        self.left = BinarySearchTreeNode(data)
    else:
      # add data in right sub tree
      if self.right:
        self.right.add_child(data)
      else:
        self.right = BinarySearchTreeNode(data)

  # in order traversal
  def in_order_traversal(self):
    elements = []
    # visit left tree
    if self.left:
      elements += self.left.in_order_traversal()

    # visit base node
    elements.append(self.data)

    # visit right node
    if self.right:
      elements += self.right.in_order_traversal()
    return elements

  # search an element
  def search(self, val):
    if self.data == val:
      return True
    if val<self.data:
      if self.left:
        return self.left.search(val)
      else:
        return False

    if self.right:
      return self.right.search(val)
    else:
      return False


def build_tree(elements):
  root = BinarySearchTreeNode(elements[0])
  for i in range(1, len(elements)):
    root.add_child(elements[i])
  return root

if __name__ == "__main__":
  nums = [17,2,4,64,34,11,32]
  num_tree = build_tree(nums)
  print(num_tree.in_order_traversal())
  print(num_tree.search(11))
````

````python
# quick sort
# hoare Partition and Lumoto Partition
def swap(a, b, arr):
  if a!=b:
    arr[a], arr[b] = arr[b], arr[a]

def partition(elements, start, end):
  pivot_index = start
  pivot = elements[pivot_index]
  start = pivot_index +1
  end= len(elements)-1
  while start<end:
    while start< len(elements) and elements[start]<= pivot:
      start +=1
    while elements[end]>pivot:
      end -=1
    if start<end:
      swap(start, end, elements)
  swap(pivot_index, end, elements)

  return end

def quick_sort(elements, start , end):
  if start<end:
    part = partition(elements, start, end)
    quick_sort(elements, start, part-1)
    quick_sort(elements, part+1, end)


if __name__ == "__main__":
  elements = [11,9,29,7,2,15,28]
  quick_sort(elements,0, len(elements)-1)
  print(elements)
````

````python
# lumoto Partition - complexity = O(nlogn) and O(n**2) in worst case

def swap(a, b, arr):
  arr[a], arr[b] = arr[b], arr[a]

def partition(elements, start=0, end=len(elements)-1):
  pivot = elements[end]
  p_index = start
  for i in range(start, end):
    if elements[i] <= pivot:
      swap(i, p_index, elements)
      p_index +=1
  swap(p_index, end, elements)
  return p_index

def quick_sort_lumoto(elements, start=0, end=len(elements)-1):
  if len(elements)==1:
    return
  if start<end:
    part = partition(elements, start, end)
    quick_sort_lumoto(elements, start, part-1)
    quick_sort_lumoto(elements, part+1, end)

if __name__ == "__main__":
  elements = [11,9,29,7,2,15,28]
  quick_sort_lumoto(elements)
  print(elements)
````

Sort the elements of a given list using shell sort, but with slight modification. Remove all the repeating occurances of elements while sorting.

Traditionally, when comparing two elements in shell sort, we swap if first element is bigger than second, and do nothing otherwise.

In this modified shell sort with duplicate removal, if values are same, we will delete one of the two elements we are comparing before starting the next pass for the reduced gap.

For example, given the unsorted list [2, 1, 5, 7, 2, 0, 5, 1, 2, 9, 5, 8, 3], after sorting using shell sort without duplicates, the sorted list would be:
```
[0, 1, 2, 3, 5, 7, 8, 9]
```

````python
def shell_sort(arr):
  size = len(arr)
  div = 2
  gap = size//div
  while gap > 0:
    index_to_delete = []
    for i in range(gap, size):
      anchor = arr[i]
      j = i
      while j >= gap and arr[j-gap] >= anchor:
        if arr[j-gap] == anchor:
          index_to_delete.append(j)
        arr[j] = arr[j-gap]
        j -= gap

      arr[j] = anchor

    index_to_delete = list(set(index_to_delete))
    index_to_delete.sort()
    if index_to_delete:
      for i in index_to_delete[-1::-1]:
        del arr[i]

    size = len(arr)
    div *= 2
    gap = size//div


if __name__ == "__main__":
  arr = [2, 1, 5, 7, 2, 0, 5, 1, 2, 9, 5, 8, 3]
  shell_sort(arr)
  print(arr)
````

````python
# given an array arr, consists of only zeros, ones and twos. sort the same array in-place and return it. don't create a new array. time complexity - O(n)

def sort_array(arr,n):
  low, mid, high = 0, 0, n-1
  while mid <= high:
    if arr[mid] ==2:
      arr[mid], arr[high] = arr[high], arr[mid]
      high -=1
    elif arr[mid] ==1:
      mid +=1
    else:
      arr[mid], arr[low] = arr[low], arr[mid]
      low +=1
      mid +=1

  return arr

sort_array([0,1,2,0,1,2],6)
````

````python
def count_vowel_consonants(s):
  vowels = 'aeiou'
  vowel_count = 0
  consonant_count = 0
  cleaned_txt = s.lower()
  for char in cleaned_txt:
    if char.isalpha():
      if char in vowels:
        vowel_count +=1
      else:
        consonant_count +=1
  return vowel_count, consonant_count

print(count_vowel_consonants('Alphabet'))
````

Build a quick example of a system showing Encapsulation using getter and setter logic.

````python
from os import access
class BankAccount:
  def __init__(self, account_number, balance):
    self.__account_number = account_number
    self.__balance =balance

  def get_account_number(self):
    return self.__account_number

  def get_balance(self):
    return self.__balance

  def set_balance(self, balance):
    if balance >=0:
      self.__balance = balance
    else:
      print("Invalid balance")

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
    else:
      print("Invalid deposit amount")

  def withdraw(self, amount):
    if amount > 0 and amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Invalid withdrawal amount or insufficient funds")

account = BankAccount("123456789", 1000)
print(account.get_account_number())
print(account.get_balance())
deposit_amount = 500
account.deposit(deposit_amount)
print(f"Deposited ${deposit_amount}. New balance: ${account.get_balance()}")
withdraw_amount = 1501
account.withdraw(withdraw_amount)
print(f"Withdrew ${withdraw_amount}. New balance: ${account.get_balance()}")
````

````python
from typing_extensions import final
def clean_file(source_file, target_file):
  try:
    with open(source_file, 'r') as source, open(target_file, 'w') as target:
      for line in source:
        line = line.strip()
        if line:
          target.write(line + '\n')
  except FileNotFoundError:
    print(f"File not found: {source_file}")
  except Exception as e:
    print(f"An error occurred: {e}")
  finally:
    print('process completed')
````
