```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
print(df.head())
```

What are the columns available?

```python
df.columns
```

What is the average sepal length?

```python
print(df['sepal_length'].mean())
df['sepal_length'].mean()
```

Which species have the longest petal?

```python
print(df.groupby('species')['petal_length'].mean())
```

```python
sns.barplot(x='species', y='petal_length', data=df)
plt.title("Average Petal Length per Species")
plt.show()
```

```python
df.plot(x = 'species', y = 'petal_length')
```

Virginica has the longest petal

Are sepal length and petal length related?

```python
df[['petal_length', 'sepal_length']].corr()
```

```python
sns.scatterplot(x= 'sepal_length', y= 'petal_length', data=df, hue= 'species')
plt.title("Sepal length Vs Petal Length ")
plt.show()
```

```python
sns.heatmap(df[['petal_length', 'sepal_length']].corr(), annot= True, cmap= 'coolwarm')
plt.title("Feature Correlation with heatmap")
plt.show()
```
